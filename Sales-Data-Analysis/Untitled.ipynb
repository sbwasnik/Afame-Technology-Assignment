# 3. Feature Engineering

## 3.1. Missing Values

```python
df.isnull().sum()
```

```python
df['Postal Code'] = df['Postal Code'].fillna(0)
```

## 3.2. Convert datetime

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

# 4. Exploratory Data Analysis

## 4.1. Total Sales

```python
total_sales = df['Sales'].sum()
total_quantity = df['Quantity'].sum()
total_profit = df['Profit'].sum()
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M')).sum(numeric_only=True)['Sales']
yearly_sales = df.groupby(df['Order Date'].dt.to_period('Y')).sum(numeric_only=True)['Sales']
```

```python
# Plot monthly sales trend
plt.figure(figsize=(14, 7))
monthly_sales.plot()
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
# Plot yearly sales trend
plt.figure(figsize=(14, 7))
yearly_sales.plot()
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

## 4.2. Best Selling Product

```python
top_products_sales = df.groupby('Product Name').sum(numeric_only=True)['Sales'].sort_values(ascending=False).head(10)
top_products_quantity = df.groupby('Product Name').sum(numeric_only=True)['Quantity'].sort_values(ascending=False).head(10)
top_products_profit = df.groupby('Product Name').sum(numeric_only=True)['Profit'].sort_values(ascending=False).head(10)
```

```python
# Plot top products by sales
plt.figure(figsize=(14, 7))
top_products_sales.plot(kind='bar')
plt.title('Top 10 Products by Sales')
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.show()
```

```python
# Plot top products by quantity
plt.figure(figsize=(14, 7))
top_products_quantity.plot(kind='bar')
plt.title('Top 10 Products by Quantity Sold')
plt.xlabel('Product Name')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
# Plot top products by profit
plt.figure(figsize=(14, 7))
top_products_profit.plot(kind='bar')
plt.title('Top 10 Products by Profit')
plt.xlabel('Product Name')
plt.ylabel('Profit')
plt.show()
```

## 4.3. Sales Distribution

```python
# Bar Plot for Average Sales by Category
avg_sales_by_category = df.groupby('Category').mean(numeric_only=True)['Sales']
plt.figure(figsize=(14, 7))
avg_sales_by_category.plot(kind='bar')
plt.title('Average Sales by Category')
plt.xlabel('Category')
plt.ylabel('Average Sales')
plt.show()
```

```python
# Bar Plot for Average Sales by Category
avg_sales_by_category = df.groupby('Region').mean(numeric_only=True)['Sales']
plt.figure(figsize=(14, 7))
avg_sales_by_category.plot(kind='bar')
plt.title('Average Sales by Region')
plt.xlabel('Region')
plt.ylabel('Average Sales')
plt.show()
```

```python
# Bar Plot for Average Sales by Customer Segment
avg_sales_by_category = df.groupby('Segment').mean(numeric_only=True)['Sales']
plt.figure(figsize=(14, 7))
avg_sales_by_category.plot(kind='bar')
plt.title('Average Sales by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Average Sales')
plt.show()
```

```python
# Histogram for Sales Distribution
plt.figure(figsize=(14, 7))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```
