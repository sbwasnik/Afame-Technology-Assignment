# 3. Feature Engineering

```python
df.columns
```

```python
# Calculate the mean of three satisfaction columns and assign the result to a new column 'PerformanceSatisfactionIndex'

df['PerformanceSatisfactionIndex'] = df[['EnvironmentSatisfaction', 'JobSatisfaction', 'RelationshipSatisfaction']].mean(axis=1)
```

```python
# Calculate the experience prior to the current job by subtracting the number of years at the current company from the total working years, and assign the result to a new column

df['ExperiencePriorToCurrentJob'] = df['TotalWorkingYears'] - df['YearsAtCompany']
```

```python
# Calculate the employee engagement score by averaging job involvement, performance rating, and work-life balance, and assign the result to a new column

df['EmployeeEngagementScore'] = (df['JobInvolvement'] + df['PerformanceRating'] + df['WorkLifeBalance']) / 3
```

## 3.1. Mapping numerical values to descriptive labels

```python
education_map = {
    1: 'Below College',
    2: 'College',
    3: 'Bachelor',
    4: 'Master',
    5: 'Doctor'
}
```

```python
environment_satisfaction_map = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}
```

```python
job_involvement_map = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}
```

```python
job_satisfaction_map = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}
```

```python
performance_rating_map = {
    1: 'Low',
    2: 'Good',
    3: 'Excellent',
    4: 'Outstanding'
}
```

```python
relationship_satisfaction_map = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}
```

```python
work_life_balance_map = {
    1: 'Bad',
    2: 'Good',
    3: 'Better',
    4: 'Best'
}
```

```python
df['Education'] = df['Education'].map(education_map)
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].map(environment_satisfaction_map)
df['JobInvolvement'] = df['JobInvolvement'].map(job_involvement_map)
df['JobSatisfaction'] = df['JobSatisfaction'].map(job_satisfaction_map)
df['PerformanceRating'] = df['PerformanceRating'].map(performance_rating_map)
df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].map(relationship_satisfaction_map)
df['WorkLifeBalance'] = df['WorkLifeBalance'].map(work_life_balance_map)
df['Overtime_Preference'] = df['OverTime'].map({'Yes': 1, 'No': 0})
```

```python
def generation(age):
    if (age >= 97) & (age <= 102):
        return 'WWII'
    elif (age >= 79) & (age <= 96):
        return 'Post War'
    elif (age >= 70) & (age <= 78):
        return 'Boomers I*'
    elif (age >= 60) & (age <= 69):
        return 'Boomers II (a/k/a Generation Jones)*'
    elif (age >= 44) & (age <= 59):
        return 'Gen X'
    elif (age >= 28 ) & (age <= 43):
        return 'Millennials'
    elif (age >= 12 ) & (age <= 27):
        return 'Gen Z'
```

```python
def distance_category(distance):
    if (distance >= 1) & (distance <= 10):
        return 'Short Distance'
    elif (distance >= 11) & (distance <= 20):
        return 'Medium Distance'
    elif (distance >= 21) & (distance <= 30):
        return 'Long Distance'
```

```python
df['AgeGeneration'] = df['Age'].apply(generation)
df['DistanceFromHomeCategory'] = df['DistanceFromHome'].apply(distance_category)
```

```python
df = df.reindex(sorted(df.columns), axis=1)
```
